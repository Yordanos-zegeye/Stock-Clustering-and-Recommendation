```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


data_path = '../results/clustered_data.csv'
stock_data = pd.read_csv(data_path)
```

```python
def visualize_clusters(data, x_feature, y_feature, cluster_column):
    """Scatter plot to visualize clusters."""
    plt.figure(figsize=(10, 6))
    sns.scatterplot(
        data=data, x=x_feature, y=y_feature, hue=cluster_column, palette='viridis', s=100
    )
    plt.title(f"{y_feature} vs {x_feature} by Cluster", fontsize=16)
    plt.xlabel(x_feature, fontsize=14)
    plt.ylabel(y_feature, fontsize=14)
    plt.legend(title='Cluster', fontsize=12)
    plt.grid(True)
    plt.show()
```

### Visualize ROE vs Current Price by cluster

```python
visualize_clusters(stock_data, 'Current Price', 'ROE', 'Cluster')
```

### Visualize P/E Ratio vs Earnings Per Share by cluster

```python
visualize_clusters(stock_data, 'Earnings Per Share', 'P/E Ratio', 'Cluster')
```

### Visualize Volatility vs ROE by cluster

```python
visualize_clusters(stock_data, 'ROE', 'Volatility', 'Cluster')
```

### Heatmap to show the average values of key financial indicators for each cluster

```python
numerical_features = ['Current Price', 'Price Change', 'Volatility', 'ROE', 'Cash Ratio',
                      'Net Cash Flow', 'Earnings Per Share', 'P/E Ratio']
cluster_means = stock_data.groupby('Cluster')[numerical_features].mean()

plt.figure(figsize=(12, 8))
sns.heatmap(cluster_means, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title("Cluster-wise Mean of Financial Indicators", fontsize=16)
plt.show()
```

### Visualize the count of stocks in each cluster

```python
plt.figure(figsize=(8, 6))
cluster_counts = stock_data['Cluster'].value_counts().sort_index()
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, palette='Blues')
plt.title("Number of Stocks in Each Cluster", fontsize=16)
plt.xlabel("Cluster", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.grid(True)
plt.show()
```
